```python
import requests
from bs4 import BeautifulSoup as bs

amueller = requests.get("https://amueller.github.io").content

soup = bs(amueller, 'html.parser')
abt_amueller = soup.find("div", "col-sm-8 text-left")
abt_amueller = abt_amueller.get_text(strip=True)
abt_amueller
```

```python
from wordcloud import WordCloud

wc_author = abt_amueller

wc = WordCloud(background_color="white", colormap="Dark2",width=500, 
               height=250, max_font_size=75, min_font_size=15,).generate(wc_author)

wc.to_file('Img/amueller_wcloud.png')
```
